# 1. Converting from One Mapping to Other

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

def img_imp_show(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    print(img.shape)
    plt.imshow(img)
    return img
```

```python
img = img_imp_show('DATA/00-puppy.jpg')
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_RGB2HLS)
plt.imshow(img)
```

# 2.Blending Images

```python
# new_pixel = alpha* pixel_1 + beta*pixel_2 + gamma
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = img_imp_show('DATA/dog_backpack.jpg')
```

```python
img2 = img_imp_show('DATA/watermark_no_copy.png')
```

### Blending Images of same size

```python
img1 = cv2.resize(img1,(1200,1200))
img2 = cv2.resize(img2,(1200,1200))
```

```python
blend = cv2.addWeighted(src1 = img1,alpha =0.5, src2=img2,beta=0.5,gamma=0)
plt.imshow(blend)
```

```python
blend = cv2.addWeighted(src1 = img1,alpha =1, src2=img2,beta=0.2,gamma=0)
plt.imshow(blend)
```

### Overlay smaller image on top of larger image (No Blending, Simpling Overlay)

```python
large_image = img_imp_show('DATA/dog_backpack.jpg')
small_image = img_imp_show('DATA/watermark_no_copy.png')
small_image = cv2.resize(small_image,(600,600))
```

```python
x_offset = 200
y_offset = 200
x_end = x_offset + small_image.shape[1]
y_end = y_offset + small_image.shape[0]
large_image[y_offset:y_end,x_offset:x_end] = small_image
```

```python
plt.imshow(large_image)
```

# 3.Image Thresholding

```python
crossword = cv2.imread('DATA/crossword.jpg',0)
plt.imshow(crossword,cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(crossword)
```

```python
thresh,img_th1 = cv2.threshold(crossword,140,255,cv2.THRESH_BINARY)
show_pic(img_th1)
```

```python
img_th2 = cv2.adaptiveThreshold(crossword,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,5,8)
show_pic(img_th2)
```

```python
blend = cv2.addWeighted(src1 = img_th1,alpha =0.5, src2=img_th2,beta=0.5,gamma=0)
show_pic(blend)
```
