# 1. Opening Image Files using OPENCV

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
# Correct Image Path
img = cv2.imread('DATA/00-puppy.jpg')
type(img),img.shape
```

```python
# Wrong Image Path
img2 = cv2.imread('DATA/00-puppyeeeee.jpg')
type(img2)
```

```python
plt.imshow(img)
# Blue because difference in channels
# MATPLOTLIN --> RGB
# OpenCV --> BGR
# Thus B and R gets swapped result in weird colors
```

```python
# Changing color channel coding
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
# Instead of all this pre-define input channel code
img_gray = cv2.imread('DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray')
```

```python
#Resizing IMAGE
fix_img.shape
```

```python
new_img = cv2.resize(fix_img,(1000,400))
plt.imshow(new_img)
new_img.shape
```

```python
#Resizing IMAGE by ratio
width_ratio = 0.5
h_ratio = 0.5

new_img = cv2.resize(fix_img,(0,0),fix_img,width_ratio,h_ratio)
plt.imshow(new_img)
new_img.shape
```

```python
#flip Image use 0 or -1 
new_img = cv2.flip(fix_img,0)
plt.imshow(new_img)
```

```python
#saving file
cv2.imwrite('naya_hai_yaha.jpg',new_img)
```

# 3.Drawing on Images - Part One - Basic Shapes

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blk = np.zeros((512,512,3),dtype=np.int16)
blk.shape
```

```python
plt.imshow(blk)
```

```python
cv2.rectangle(blk,pt1=(384,0),pt2=(510,150),color = (0,255,0),thickness=10)
plt.imshow(blk)
```

```python
cv2.rectangle(blk,pt1=(200,300),pt2=(300,200),color = (0,255,0),thickness=5)
plt.imshow(blk)
```

```python
cv2.circle(blk,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(blk)
```

```python
cv2.circle(blk,center=(100,100),radius=50,color=(255,0,0),thickness=-8)
plt.imshow(blk)
```

```python
cv2.line(blk,pt1=(0,0),pt2=(512,512),color = (102,255,255),thickness=5)
plt.imshow(blk)
```

# 4.Drawing on Images Part Two - Text and Polygons

```python
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(blk,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType = cv2.LINE_AA)
plt.imshow(blk)
```

# 4. Text And Polygon

```python
blk_2 = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(blk_2)
```

```python
vert = np.array([ [100,300], [200,200], [400,300], [200,400]],np.int32)
vert.shape
# It needs to be in 3d for each color channel
```

```python
pts = vert.reshape((-1,1,2))
pts.shape
```

```python
cv2.polylines(blk_2,[pts],isClosed=True,color=[255,0,0],thickness=3)
plt.imshow(blk_2)
```
